```python
import torch
from torch import nn
import matplotlib.pyplot as plt
```

```python
  weight = 0.7
  bias = 0.3

  start = 0
  end = 1
  step = 0.02

  X = torch.arange(start, end, step).unsqueeze(dim=1)
  y = weight * X + bias
  X[:5], y[:5]
```

```python
train_split = int(len(X) * 0.8)
X_train, y_train = X[:train_split], y[:train_split]
X_test, y_test = X[train_split:], y[train_split:]
len(X_train), len(y_train)
```

```python
def plot_prediction(X_train, y_train, X_test, y_test, y_predict):
  plt.figure(figsize=(10, 7))
  plt.scatter(X_train, y_train, c='b', s=4, label='Train Set')
  plt.scatter(X_test, y_test, c='g', s=4, label='Test Set')
  if y_predict is not None:
    plt.scatter(X_test, y_predict, c='r', s=4, label='Predict Set')
  plt.legend(prop={'size': 14})
  plt.show
```

```python
plot_prediction(X_train, y_train, X_test, y_test, None)
```

```python
from torch import nn

class LinearRegressionModel(nn.Module):
  def __init__(self):
    super().__init__()
    self.weights = nn.Parameter(torch.randn(1, requires_gred=True, dtype=torch.float))
    self.bias = nn.Parameter(torch.randn(1, requires_gred=True, dtype=torch.float))

  def forward(self, x: torch.Tensor) -> torch.Tensor:
    return self.weights * x + self.bias
```
